```python
import urllib.request
import numpy as np
import cftime
import os
from datetime import datetime, timedelta
from concurrent.futures import ThreadPoolExecutor, as_completed
from netCDF4 import Dataset, num2date,date2num
import pandas as pd
from tqdm import tqdm

if not os.path.exists(download_GPM_directory):
    os.makedirs(download_GPM_directory)
if not os.path.exists(nc_output_directory):
    os.makedirs(nc_output_directory)

# User credentials for accessing the data server
# https://www.earthdata.nasa.gov/eosdis/science-system-description/eosdis-components/earthdata-login
# you need an NASA Earth Data account (see above)
username = "diyali189"
password = "e55kVb7$di2vjv3"

# Initialize global variables to store precipitation data, latitude, longitude, and times
precipitation_data = []
latitude = None
longitude = None
times = []

def create_url(year, month, day):
    """
    Create a URL for downloading data for a specific year, month, and day.
    """
    base_url = "https://gpm1.gesdisc.eosdis.nasa.gov/data/GPM_L3/GPM_3IMERGDF.07/"
    date_str = f"{year}/{month:02d}/3B-DAY.MS.MRG.3IMERG.{year}{month:02d}{day:02d}-S000000-E235959.V07B.nc4"
    return base_url + date_str

def create_fname(year, month, day):
    fname = f"3B-DAY.MS.MRG.3IMERG.{year}{month:02d}{day:02d}-S000000-E235959.V07B.nc4"
    return fname

def download_and_process(year, month, day, retries=3):
    """
    Download and process the data for a given date.
    """
    url = create_url(year, month, day)
    fname = create_fname(year, month, day)
    #file_path = os.path.join(download_GPM_directory, f'temp_{year}_{month:02d}_{day:02d}.nc')
    file_path = os.path.join(download_GPM_directory, fname)
    attempts = 0

    while attempts <= retries:
        try:
            urllib.request.urlretrieve(url, file_path)
            print(f"File for {year}-{month:02d}-{day:02d} downloaded successfully")
            if os.path.exists(file_path) and os.path.getsize(file_path) > 0:
                print(f"Processing file: {file_path}")
                process_nc_file(file_path, "precipitation")
                return
            else:
                print(f"File {file_path} does not exist or is empty.")
                attempts += 1
        except Exception as e:
            print(f"Error for {year}-{month:02d}-{day:02d}: {e}. Attempt {attempts+1} of {retries}")
            attempts += 1

    print(f"Failed to download and process file after {retries} retries: {year}-{month:02d}-{day:02d}")

def process_nc_file(nc_file, variable_name, time_index=0, lat_min=None, lat_max=None, lon_min=None, lon_max=None):
    """
    Process the .nc (NetCDF) file to extract precipitation data.
    """
    global precipitation_data, latitude, longitude, times
    try:
        with Dataset(nc_file, 'r') as dataset:
            data = dataset.variables[variable_name][time_index, :, :]
            data = np.transpose(data)
            fill_value = dataset.variables[variable_name]._FillValue
            data = np.ma.masked_equal(data, fill_value)
            data = np.ma.filled(data, np.nan)
            
            if latitude is None or longitude is None:
                latitude = dataset.variables['lat'][:]
                longitude = dataset.variables['lon'][:]

            precipitation_data.append(data[np.newaxis, ...])
            times.append(dataset.variables['time'][time_index])
    except Exception as e:
        print(f"Error processing file: {e}")

def remove_duplicate_entries():
    """
    Remove duplicate entries from the global precipitation data.
    """
    global times, precipitation_data
    unique_times = []
    unique_precipitation_data = []

    for time, data in zip(times, precipitation_data):
        if time not in unique_times:
            unique_times.append(time)
            unique_precipitation_data.append(data)

    times, precipitation_data = unique_times, unique_precipitation_data        

def save_precipitation_data_to_nc(filename, lat_min, lat_max, lon_min, lon_max):
    global precipitation_data, latitude, longitude, times

    if not precipitation_data or latitude is None or longitude is None:
        print("No precipitation data available to save.")
        return

    time_units = "days since 1980-01-06 00:00:00"
    cftime_dates = num2date(times, units=time_units, calendar='gregorian')
    py_dates = [pd.to_datetime(str(date)) for date in cftime_dates]
    sorted_indices = np.argsort(py_dates)
    times_sorted = np.array(cftime_dates)[sorted_indices]
    precipitation_data_sorted = [precipitation_data[i] for i in sorted_indices]

    # Convert sorted datetime objects back to numeric time values for netCDF storage
    numeric_times = date2num(times_sorted, units=time_units, calendar='gregorian')

    lat_indices = np.where((latitude >= lat_min) & (latitude <= lat_max))
    lon_indices = np.where((longitude >= lon_min) & (longitude <= lon_max))
    filtered_latitude = latitude[lat_indices]
    filtered_longitude = longitude[lon_indices]
    filtered_precipitation_data = [data[:, lat_indices[0], :][:, :, lon_indices[0]] for data in precipitation_data_sorted]

    with Dataset(filename, 'w', format='NETCDF4') as ncfile:
        ncfile.createDimension('time', len(numeric_times))
        ncfile.createDimension('lat', len(filtered_latitude))
        ncfile.createDimension('lon', len(filtered_longitude))

        times_nc = ncfile.createVariable('time', 'f4', ('time',))
        latitudes_nc = ncfile.createVariable('lat', 'f4', ('lat',))
        longitudes_nc = ncfile.createVariable('lon', 'f4', ('lon',))
        precip_nc = ncfile.createVariable('precipitation', 'f4', ('time', 'lat', 'lon',), zlib=True)

        times_nc[:] = numeric_times
        latitudes_nc[:] = filtered_latitude
        longitudes_nc[:] = filtered_longitude
        precip_nc[:, :, :] = filtered_precipitation_data

        latitudes_nc.units = "degrees_north"
        longitudes_nc.units = "degrees_east"
        precip_nc.units = "mm/day"
        precip_nc.long_name = "Daily mean precipitation rate (combined microwave-IR) estimate."
        precip_nc.coordinates = "time lon lat"
        precip_nc._ChunkSizes = [1, len(filtered_latitude), len(filtered_longitude)]

        ncfile.source = "GPM IMERG Final Precipitation L3 1 day 0.1 degree x 0.1 degree V07 DOI: 10.5067/GPM/IMERGDF/DAY/07"
        times_nc.units = time_units

        print(f"Data with attributes saved to {filename}")


def download_and_process_concurrently(start_date_str, end_date_str, max_workers=1):
    """
    Download and process data concurrently for a range of dates with a progress bar.
    """
    start_date = datetime.strptime(start_date_str, '%Y-%m-%d')
    end_date = datetime.strptime(end_date_str, '%Y-%m-%d')
    delta = end_date - start_date
    failed_downloads = []

    # Initialize tqdm progress bar
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        date_range = [(start_date + timedelta(days=i)) for i in range(delta.days + 1)]
        
        with tqdm(total=len(date_range), desc="Downloading and Processing Data", unit="file") as pbar:
            for current_date in date_range:
                futures.append(executor.submit(download_and_process, current_date.year, current_date.month, current_date.day))

            for future in as_completed(futures):
                result = future.result()  # Wait for the future to complete
                if result is not None:
                    failed_downloads.append(result)
                pbar.update(1)  # Update progress bar for each completed task

    if failed_downloads:
        print(f"Failed downloads: {failed_downloads}")

    return start_date, end_date

# Set up user authentication handler
password_manager = urllib.request.HTTPPasswordMgrWithDefaultRealm()
password_manager.add_password(None, "https://urs.earthdata.nasa.gov", username, password)
auth_handler = urllib.request.HTTPBasicAuthHandler(password_manager)
opener = urllib.request.build_opener(urllib.request.HTTPRedirectHandler(), urllib.request.HTTPCookieProcessor(), auth_handler)
urllib.request.install_opener(opener)

# Example execution of the script
lat_min, lat_max = 31.85, 37.70  # Latitude boundaries
lon_min, lon_max = -85.35, -75.55  # Longitude boundaries

start_date, end_date = download_and_process_concurrently(start_date, end_date)

remove_duplicate_entries()

# Modify the filename to save in the final_dir
filename = os.path.join(nc_output_directory, f"GPM_precip_IMERGDFv07_{start_date.strftime('%Y%m%d')}{end_date.strftime('%Y%m%d')}.nc")
save_precipitation_data_to_nc(filename, lat_min, lat_max, lon_min, lon_max)
```

**Merge_GPM_Grid_nc**

```python
import os

merged_output_directory = '/work/sds-lab/Diyali/NASA_storm/Code_Output/GPM+Precip_Data/Merged'

if not os.path.exists(merged_output_directory):
    os.makedirs(merged_output_directory)
```

```python
import os
from netCDF4 import Dataset, num2date, date2num
import numpy as np

def merge_nc_files(nc_output_directory, merged_output_directory, data_product_name, product_id):
    nc_files = [os.path.join(nc_output_directory, file) for file in os.listdir(nc_output_directory) if file.endswith('.nc')]
    if not nc_files:
        raise FileNotFoundError("No .nc files found in the specified directory.")

    base_file = nc_files[0]
    additional_files = nc_files[1:]

    with Dataset(base_file, 'r') as base_nc:
        base_vars = {var: base_nc.variables[var][:] for var in base_nc.variables if var not in ['time', 'lat', 'lon']}
        base_time = base_nc.variables['time']
        units = base_time.units
        calendar = getattr(base_time, 'calendar', 'standard')

        aggregated_dates = [num2date(base_time[:], units, calendar)]
        aggregated_data = {var: [data] for var, data in base_vars.items()}

        for file_path in additional_files:
            with Dataset(file_path, 'r') as nc:
                time = nc.variables['time']
                add_calendar = getattr(time, 'calendar', calendar)  # Use the base calendar as default
                dates = num2date(time[:], time.units, add_calendar)
                aggregated_dates.append(dates)

                for var, data in aggregated_data.items():
                    if var in nc.variables:
                        data.append(nc.variables[var][:])

        aggregated_dates = np.concatenate(aggregated_dates)
        date_sort_idx = np.argsort(aggregated_dates)
        sorted_dates = aggregated_dates[date_sort_idx]
        #filename_date_range = f"{sorted_dates[0].strftime('%Y%m%d')}_{sorted_dates[-1].strftime('%Y%m%d')}.nc"
        filename_date_range = f"{sorted_dates[0].strftime('%Y%m%d')}{sorted_dates[-1].strftime('%Y%m%d')}.nc"
        output_file = os.path.join(merged_output_directory, f"Merged_GPM_precip_{product_id}_{filename_date_range}")

        with Dataset(output_file, 'w', format='NETCDF4') as new_nc:
            new_nc.description = 'Merged NetCDF data'
            new_nc.data_product = data_product_name
            new_nc.product_id = product_id

            new_nc.createDimension('time', None)
            new_nc.createDimension('lat', len(base_nc.dimensions['lat']))
            new_nc.createDimension('lon', len(base_nc.dimensions['lon']))

            time_var = new_nc.createVariable('time', 'f8', ('time',))
            time_var[:] = date2num(sorted_dates, units, calendar)
            time_var.units = units
            time_var.calendar = calendar

            lat_var = new_nc.createVariable('lat', 'f4', ('lat',))
            lon_var = new_nc.createVariable('lon', 'f4', ('lon',))
            lat_var[:] = base_nc.variables['lat'][:]
            lon_var[:] = base_nc.variables['lon'][:]

            for var, data in aggregated_data.items():
                dtype = base_nc.variables[var].datatype
                fill_value = getattr(base_nc.variables[var], '_FillValue', None)
                var_data = np.concatenate(data)[date_sort_idx]
                new_var = new_nc.createVariable(var, dtype, ('time', 'lat', 'lon'), fill_value=fill_value)
                new_var[:] = var_data
                for attr in base_nc.variables[var].ncattrs():
                    setattr(new_var, attr, getattr(base_nc.variables[var], attr))

data_product_name = "GPM"
product_id = "IMERGDFv07"
merge_nc_files(nc_output_directory, merged_output_directory, data_product_name, product_id)
```

**Precip_Catchments_spatial_average_nc**

```python
import os
import xarray as xr
import geopandas as gpd
import pandas as pd
import xagg as xa
import rioxarray
from datetime import datetime  
import warnings

warnings.filterwarnings('ignore')

def aggregate_data_to_csv(nc_input_file, shp_path, rivid_path, csv_output_directory):
    # Extract date range from the filename
    base_name = os.path.basename(nc_input_file)
    date_str = base_name.split('.')[0].split('_')[-1]
    range_start, range_end = date_str[:8], date_str[8:]
    output_csv_name = f'GPM_precip_IMERGDFv07_SA_catchments_{range_start}_{range_end}.csv'
    formatted_output_csv_path = os.path.join(csv_output_directory, output_csv_name)

    # Load the NetCDF data
    data_precep = xr.open_dataset(nc_input_file)
    data_precep = data_precep.sortby('time')  # Sort by time

    data_precep = data_precep.assign_coords(lat=data_precep.lat, lon=data_precep.lon)

    # Load and prepare shapefile data
    data2 = gpd.read_file(shp_path)
    if data_precep.rio.crs is None:
        data_precep.rio.write_crs("epsg:4326", inplace=True)
    if data2.crs is None:
        data2.set_crs("epsg:4326", inplace=True)
    data2 = data2.to_crs(data_precep.rio.crs.to_string())

    # Load River ID data
    data_rivid = pd.read_table(rivid_path)
    data_rivid['COMID'] = data_rivid['COMID'].astype(int)
    data_rivid['RIVID'] = data_rivid['RIVID'].astype(int)

    # Merge shapefile data with RIVID data
    data2 = data2.merge(data_rivid, on='COMID', how='left')

    # Create xagg weightmap
    try:
        weightmap = xa.pixel_overlaps(data_precep, data2)
        print("Weightmap created successfully!")
    except Exception as e:
        print(f"Error creating weightmap: {e}")

    # Aggregate temperature data
    try:
        variable = 'precipitation'  # Adjust this variable name as per your dataset
        aggregated = xa.aggregate(data_precep[variable], weightmap)
        print("Aggregation successful!")
    except Exception as e:
        print(f"Error during aggregation: {e}")

    # Convert aggregated data to DataFrame
    aggregated_df = aggregated.to_dataframe().reset_index()

    # Sort DataFrame by RIVID and then by COMID
    aggregated_df.sort_values(by=['RIVID', 'COMID'], inplace=True)

    # Pivot the DataFrame
    pivot_df = aggregated_df.pivot_table(index=['RIVID', 'COMID'], columns='time', values=variable, aggfunc='mean').reset_index()

    # Format the date columns
    pivot_df.columns = [col.strftime('%Y-%m-%d') if isinstance(col, datetime) else col for col in pivot_df.columns]

    # Ensure the pivot table is sorted correctly by RIVID and then by COMID
    pivot_df.sort_values(by=['RIVID', 'COMID'], inplace=True)

    # Save to CSV
    pivot_df.to_csv(formatted_output_csv_path, index=False)
    print(f"Data saved to {formatted_output_csv_path}")

aggregate_data_to_csv(nc_input_file, path_dbf_catchment, path_rivid, csv_cat_output_directory)
```

**Precip_Counties_spatial_average_nc**

```python
import os
import xarray as xr
import geopandas as gpd
import pandas as pd
import xagg as xa
import rioxarray
from datetime import datetime
import warnings

warnings.filterwarnings('ignore')

def aggregate_data_to_csv(nc_input_file, path_shp_county, csv_output_directory):
    # Extract date range from the filename
    base_name = os.path.basename(nc_input_file)

    # Assuming filename format is 'YYYYMMDDYYYYMMDD.nc'
    date_str = base_name.split('.')[0]  # Split on '.' and take the first part
    range_start, range_end = date_str[-16:-8], date_str[-8:]  # Last 16 characters without extension
    output_csv_name = f'GPM_precip_IMERGDFv07_SA_Counties_{range_start}_{range_end}.csv'
    
    formatted_output_csv_path =  os.path.join(csv_output_directory, output_csv_name)
    
    # Load the NetCDF data
    ds_temp = xr.open_dataset(nc_input_file)
    ds_temp = ds_temp.sortby('time')  # Sort by time if necessary

    # Ensure that 'lat' and 'lon' are indexed
    ds_temp = ds_temp.set_index({'lat': 'lat', 'lon': 'lon'})

    # Load and prepare shapefile data
    data2 = gpd.read_file(path_shp_county)
    if ds_temp.rio.crs is None:
        ds_temp.rio.write_crs("epsg:4326", inplace=True)
    data2 = data2.to_crs(ds_temp.rio.crs.to_string())

    # Create xagg weightmap
    weightmap = xa.pixel_overlaps(ds_temp, data2)

    # Aggregate temperature data assuming 'LST_Day' is the variable name in your dataset
    variable = 'precipitation' # Adjust this variable name as per your dataset
    aggregated = xa.aggregate(ds_temp[variable], weightmap)

    # Convert the aggregated data to a DataFrame and reset index
    aggregated_df = aggregated.to_dataframe().reset_index()

    # Check if 'GEOID' is in columns
    if 'GEOID' not in aggregated_df.columns:
        print("GEOID column is missing in the DataFrame. Please check your data sources.")
        return  # Exit the function if GEOID is missing

    # Convert all relevant columns to numeric values, coercing errors
    for col in aggregated_df.columns:
        if col != 'GEOID' and col != 'time':  # Exclude 'GEOID' and 'time' from conversion
            aggregated_df[col] = pd.to_numeric(aggregated_df[col], errors='coerce')

    # Group by GEOID and time, and aggregate the values to ensure uniqueness
    aggregated_df = aggregated_df.groupby(['GEOID', 'time']).mean().reset_index()

    # Pivot the DataFrame to get dates as columns and 'LST_Day' values as cells
    pivot_df = aggregated_df.pivot(index='GEOID', columns='time', values=variable)

    # Optionally fill NaN values with an identifier or leave as is if handling of NaN is required later
    pivot_df = pivot_df.fillna('No data')  # or `pivot_df.fillna(0)` if you prefer zeros

    # Save the pivoted DataFrame to CSV
    pivot_df.to_csv(formatted_output_csv_path)

    print(f'Pivoted data saved to CSV at {formatted_output_csv_path}')

aggregate_data_to_csv(nc_input_file, path_shp_county, csv_co_output_directory)
```

Merge_Precip_Catchment_csv.ipynb

```python
import pandas as pd
import os
import re
from datetime import datetime

def merge_csv_files(csv_cat_output_directory, merged_output_directory, output_file_prefix):
    csv_files = [os.path.join(csv_cat_output_directory, f) for f in os.listdir(csv_cat_output_directory) if f.endswith('.csv')]
    
    if not csv_files:
        raise FileNotFoundError("No CSV files found in the specified folder.")
    
    # Initialize a list to hold DataFrames
    dataframes = []
    
    for csv_file in csv_files:
        df = pd.read_csv(csv_file)
        # Check if both necessary columns exist in the DataFrame
        if 'COMID' not in df.columns or 'RIVID' not in df.columns:
            print(f"Warning: Skipping file {csv_file} as it does not contain 'COMID' and/or 'RIVID'.")
            continue  # Skip this file
        
        dataframes.append(df)

    # Start merging from the first DataFrame
    merged_df = dataframes[0]

    # Merge each remaining DataFrame
    for df in dataframes[1:]:
        merged_df = pd.merge(merged_df, df, on=['COMID', 'RIVID'], how='left')

    # Identify and sort date columns
    date_columns = [col for col in merged_df.columns if re.match(r'\d{4}-\d{2}-\d{2}', col)]
    non_date_columns = [col for col in merged_df.columns if col not in date_columns]

    # Sort date columns chronologically
    date_columns_sorted = sorted(date_columns, key=lambda x: datetime.strptime(x, '%Y-%m-%d'))

    # Update merged_df to use sorted date columns
    merged_df = merged_df[non_date_columns + date_columns_sorted]

    # Determine the start date and end date from sorted date columns
    if date_columns_sorted:
        start_date = date_columns_sorted[0]
        end_date = date_columns_sorted[-1]
        output_csv_file_name = f"{output_file_prefix}_{start_date}_{end_date}.csv"
    else:
        output_csv_file_name = f"{output_file_prefix}.csv"

    # Define full path for output file
    output_csv_file_path = os.path.join(merged_output_directory, output_csv_file_name)
    
    # Save the merged DataFrame to the output CSV file
    merged_df.to_csv(output_csv_file_path, index=False)

    # Print the success message
    print(f"Merged CSV file saved at: {output_csv_file_path}")

# Specify your output directory here
output_file_prefix = 'Merged_GPM_IMERGDFv07_Catchments_SA'  # Update as needed
merge_csv_files(csv_cat_output_directory, merged_output_directory, output_file_prefix)
```

Merge_Precip_Counties_csv.ipynb

```python
 import pandas as pd
import os
import re
import warnings
warnings.filterwarnings('ignore')

def merge_csv_files(csv_co_output_directory, output_file_prefix, merged_output_directory):
    csv_files = [os.path.join(csv_co_output_directory, f) for f in os.listdir(csv_co_output_directory) if f.endswith('.csv')]
    
    if not csv_files:
        raise FileNotFoundError("No CSV files found in the specified folder.")

    dfs = []
    for file_path in csv_files:
        # Load each file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Ensure column names are recognized as dates and sorted
        df.columns = pd.to_datetime(df.columns, errors='ignore')
        df = df.sort_index(axis=1)

        dfs.append(df)
    
    # Merge dataframes on 'GEOID'
    merged_df = pd.concat(dfs).groupby('GEOID').agg('max').reset_index()

    # Sort columns by date, excluding the 'GEOID' which should remain first
    date_columns = merged_df.columns.drop('GEOID')
    sorted_columns = ['GEOID'] + sorted(date_columns, key=pd.to_datetime)
    merged_df = merged_df[sorted_columns]

    # Determine the full date range for the filename
    if not date_columns.empty:
        start_date = pd.to_datetime(date_columns.min()).strftime('%Y-%m-%d')
        end_date = pd.to_datetime(date_columns.max()).strftime('%Y-%m-%d')
        output_csv_file_name = f"{output_file_prefix}_{start_date}_{end_date}.csv"
        output_csv_file_path = os.path.join(merged_output_directory, output_csv_file_name)
        merged_df.to_csv(output_csv_file_path, index=False)
        print(f"Merged CSV file saved at: {output_csv_file_path}")
    else:
        raise ValueError("Could not determine the full date range from the data.")

# Example usage

output_file_prefix = 'merged_GPM_IMERGDFv07_Counties_SA'

merge_csv_files(csv_co_output_directory, output_file_prefix, merged_output_directory)
```
