```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

county_df_95 = pd.read_csv("/home/goswami.di/Storm/county_95.csv")

# Load the county-wise 90th percentile precipitation dataset 
df = county_df_95.copy()  

# Pivot the DataFrame to get counties as rows and years as columns
df_pivot = df.pivot(index="County", columns="Year", values="95th Percentile").fillna(0)
```

```python
# Convert column names (Years) to strings
df_pivot.columns = df_pivot.columns.astype(str)

# Standardize the data (important for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pivot)

# Apply K-Means clustering
num_clusters = 3  # You can adjust this number based on the Elbow/Silhouette analysis
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df_pivot["Cluster"] = kmeans.fit_predict(X_scaled)

# Merge cluster labels back to original county DataFrame
df_clustered = df.merge(df_pivot["Cluster"], on="County")

# Display sample results
print(df_clustered.head())
```

```python
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pivot)

# Calculate WCSS for different values of k
wcss = []
K_range = range(1, 11)  # Testing clusters from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)  # Inertia = WCSS

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.xticks(K_range)
plt.show()
```

```python
# Visualize the clusters
plt.figure(figsize=(12, 6))
sns.boxplot(x="Cluster", y="95th Percentile", data=df_clustered)
plt.xlabel("Cluster")
plt.ylabel("95th Percentile Precipitation (mm/day)")
plt.title("K-Means Clustering of North Carolina Counties by 95th Percentile Precipitation")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Convert column names (Years) to strings to avoid sklearn errors
df_pivot.columns = df_pivot.columns.astype(str)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pivot)

# Apply K-Means with k=4
kmeans = KMeans(n_clusters=4, random_state=42)
df_pivot["Cluster"] = kmeans.fit_predict(X_scaled)

# Merge cluster labels back to the original county DataFrame
df_clustered = df.merge(df_pivot["Cluster"], on="County")

# Count how many times each county is assigned to each cluster
county_cluster_counts = df_clustered.groupby(["County", "Cluster"]).size().reset_index(name="Count")

# Display the most frequently assigned clusters for each county
most_common_clusters = county_cluster_counts.sort_values(by="Count", ascending=False)
```

```python
most_common_clusters
```

```python
# Count how many times each county is assigned to each cluster
county_cluster_counts = df_clustered.groupby(["County", "Cluster"]).size().reset_index(name="Count")

# Get the total number of times each county appears
county_total_counts = df_clustered["County"].value_counts().reset_index()
county_total_counts.columns = ["County", "Total_Count"]

# Merge total count with cluster assignment counts
county_cluster_counts = county_cluster_counts.merge(county_total_counts, on="County")

# Find the most frequently assigned cluster for each county
most_frequent_cluster = county_cluster_counts.loc[
    county_cluster_counts.groupby("County")["Count"].idxmax()
]

# Calculate the proportion of times each county was assigned to its most frequent cluster
most_frequent_cluster["Proportion"] = most_frequent_cluster["Count"] / most_frequent_cluster["Total_Count"]
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile of NC counties
shapefile_path = "/work/sds-lab/Diyali/NASA_storm/shapefiles/NC_counties.shp"  # Update with actual path
nc_counties_gdf = gpd.read_file(shapefile_path)

# Ensure the column names match
print("Columns in nc_counties_gdf:", nc_counties_gdf.columns)
print("Columns in most_frequent_cluster:", most_frequent_cluster.columns)

# Rename 'NAME' to 'County' in shapefile DataFrame if needed
if "NAME" in nc_counties_gdf.columns:
    nc_counties_gdf.rename(columns={"NAME": "County"}, inplace=True)

# Merge the cluster data with the shapefile using the county name
nc_counties_gdf = nc_counties_gdf.merge(most_frequent_cluster, on="County", how="left")

# Plot the clustered counties
fig, ax = plt.subplots(figsize=(12, 8))
nc_counties_gdf.plot(column="Cluster", cmap="viridis", edgecolor="black", linewidth=0.6, legend=True, ax=ax)

# Annotate county names and cluster numbers
for idx, row in nc_counties_gdf.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, f"{row['County']}\nC{int(row['Cluster'])}",
             fontsize=6, ha='center', va='center', color="black", bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

# Map title and labels
plt.title("North Carolina Counties Clustered by 95th Percentile Precipitation", fontsize=14)
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.show()
```
